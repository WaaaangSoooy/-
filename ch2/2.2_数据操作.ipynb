## 改变形状
view是Pytorch中最常用的改变tensor形状的函数，但要求数据在内存中连续存储，且变形前后元素个数必须相同  
item()，可以将一个标量Tensor转换成一个Python number


## 广播机制
当两个形状不同的Tensor按元素运算时，可能会触发广播机制：先适当复制元素使这两个Tensor形状相同都再按元素运算。
例如：

```python
x = torch.arange(1,3).view(1,2)
print(x)
y = torch.arange(1,4).view(3,1)
print(y)
print(x+y)
```
